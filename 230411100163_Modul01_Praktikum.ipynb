# Soal 1
Terdapat suatu list data berisikan nilai integer. Buatlah fungsi untuk mencari dua indeks berbeda yaitu indeks i dan j , sedemikan hingga perkalian antara (data[i]−1) ∗ (data[j]−1) menghasilkan nilai
maksimal. Return value adalah indeks i dan indeks j.

```python
def mencariIndeks(data):
    perkalian = 0
    indeks = [0,0]
    for i in range(len(data)):
        for j in range(i + 1, len(data)):
            nilai = (data[i]-1) * (data[j]-1)
            if nilai > perkalian:
                perkalian = nilai
                indeks = [i, j]

    return indeks

data = [1,5,9,10,8]
hasil_indeks = mencariIndeks(data)
hasil_perkalian = (data[hasil_indeks[0]]-1)*(data[hasil_indeks[1]]-1)
print(f"input : data = {data}")
print(f"output : {mencariIndeks(data)}")
print(f"penjelasan : data pada indeks ke-{hasil_indeks[0]} dan indeks ke-{hasil_indeks[-1]} menghasilkan hasil perkalian maksimal, yaitu {(data[hasil_indeks[0]]-1)} x {(data[hasil_indeks[1]]-1)} = {hasil_perkalian}")
```

#Soal 2
Terdapat data integer. Buatlah Fungsi untuk melakukan pengecekan apakah data tersebut adalah data angka palindrome, yaitu ketika data tersebut dibaca dari kanan atau dari kiri memiliki nilai sama.
Return value adalah True jika palindrome dan False jika sebaliknya.

```python
def cekPalindrom(angka):
  data = str(angka)
  if data[::1] == data[::-1]:
    return True
  else:
    return False

angka = int(input("input : "))
print(f"output : {cekPalindrom(angka)}")
```
